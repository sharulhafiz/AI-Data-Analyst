```python
# Correlation matrix between numerical features with target SCORE_AR
correlation = data.corr()['SCORE_AR'].sort_values(ascending=False)
top_correlation = correlation.head(15)

# Plot correlation matrix heatmap
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 10))

# Create a DataFrame from the top correlations for heatmap
top_corr_matrix = data[top_correlation.index].corr()
sns.heatmap(top_corr_matrix, annot=True, fmt=".2f")
plt.show()
```
